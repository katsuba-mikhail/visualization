````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

## Отображение легенды

````python
x = np.arange(10)
y1 = np.random.permutation(10)
y2 = np.random.permutation(10)

plt.plot(x, y1, 'x--r', label='line 1')
plt.plot(x, y2, 'g:D', label='line 2')
plt.legend()
````

````python
x = np.arange(10)
y1 = np.random.permutation(10)
y2 = np.random.permutation(10)

plt.plot(x, y1, 'x--r')
plt.plot(x, y2, 'g:D')
plt.legend(['line 1', 'line 2'])
````

## Расположение легенды на графике

````python
locs = ['best', 'upper right', 'upper left', 'lower left', 'lower right', 
        'right', 'center left', 'center right', 'lower center', 'upper center',
        'center']

x = np.arange(10)
y1 = np.random.permutation(10)
y2 = np.random.permutation(10)

plt.figure(figsize=(12, 12))        

for i in range(3):
    for j in range(4):
        if i*4+j < 11:
            plt.subplot(3, 4, i*4+j+1)
            plt.title(locs[i*4+j])
            plt.plot(x, y1, 'x--r', label='line 1')
            plt.plot(x, y2, 'c:D', label='line 2')
            plt.legend(loc=locs[i*4+j])
        else:
            break
````

Для более гибкого управление положением объекта можно воспользоваться параметром `bbox_to_anchor` функции `legend()`. Этому параметру присваивается кортеж, состоящий из четырех или двух элементов:
```
bbox_to_anchor = (x, y, width, height)
bbox_to_anchor = (x, y)
```

````python
x = np.arange(10)
y1 = np.random.permutation(10)
y2 = np.random.permutation(10)

plt.plot(x, y1, 'c:D', label='line 1')
plt.plot(x, y2, 'm--^', label='line 2')
plt.legend(bbox_to_anchor=(1, 0.6))
````

````python
x = np.arange(10)
y1 = np.random.permutation(10)
y2 = np.random.permutation(10)

plt.plot(x, y1, 'c:D', label='line 1')
plt.plot(x, y2, 'd--b', label='line 2')
plt.legend(fontsize=10, 
           shadow=True,
           framealpha=0.5,
           facecolor='y',
           edgecolor='r',
           title='Legend',
           title_fontsize=14)
````
